### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv("chipotle.tsv", delimiter="\t", 
                    dtype={'order_id':int, 'quantity':int,
                          'item_name':str, 'choice_description':str,
                          'item_price':str})
```

```python
chipo.head()
```

```python
chipo.dtypes
```

### Step 4. How many products cost more than $10.00?

```python
chipo.item_price = chipo.item_price.apply(lambda price: float(price.split("$")[1]))
chipo.dtypes
```

```python
more_than_10_count = chipo[chipo.item_price>10.].item_price.count()
more_than_10_count
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
subset = chipo[["item_name","item_price"]].drop_duplicates()
subset
```

### Step 6. Sort by the name of the item

```python
chipo = chipo.sort_values(["item_name"],ascending=[True])
chipo.head(200)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
quantity = chipo[chipo.item_price==chipo.item_price.max()]["quantity"]
quantity
```

### Step 8. How many times were a Veggie Salad Bowl ordered?

```python
quantity = chipo[chipo.item_name=="Veggie Salad Bowl"].count().quantity
quantity
```

### Step 9. How many times people orderd more than one Canned Soda?

```python
quantity = chipo[(chipo.item_name=="Canned Soda")&(chipo.quantity>1)].count().quantity
quantity
```
